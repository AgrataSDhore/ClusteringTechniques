```python
# performing the outlier capping to reduce the imapct of Extreeme Outliers
```

```python
for col in df.columns:
  if df[col].dtype!='object':
    c1=df[col].quantile(0.01)
    c2=df[col].quantile(0.99)
    df[col][df[col]<c1]=c1
    df[col][df[col]>c2]=c2
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scaled=sc.fit_transform(df2)
df_scaled
```

```python
df_scaled=pd.DataFrame(df_scaled,columns=[df2.columns])
```

```python
df_scaled
```

```python
# SSD  - sum squared Distances
# WCSS -  within the cluster sum of squared distances
# Intertia values
```

```python
# Performing Elbow method for finding optimal value of K
K_value=[2,3,4,5,6,7]
```

```python
from sklearn.cluster import KMeans
```

```python
ssd=[]
for k in K_value:
  km=KMeans(n_clusters=k,max_iter=150,random_state=45)
  km.fit(df_scaled)
  ssd.append(km.inertia_)
```

```python
K_value
```

```python
ssd
```

```python
# Plotting the Elbow plot
plt.plot(K_value,ssd)
plt.xlabel("K values")
plt.ylabel("SSD")
```

```python
# FInal Model with Number of cluster as 3 i.e K=3
kmodel=KMeans(n_clusters=3,max_iter=150,random_state=45)
kmodel.fit(df_scaled)
```

```python
kmodel.labels_
```

```python
df["Cluster ID"]=kmodel.labels_
```
